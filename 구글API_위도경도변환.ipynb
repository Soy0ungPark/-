```python
#! pip install -U googlemaps

#구글맵 api 로드
import googlemaps
from datetime import datetime


my_key = "구글 지오코딩 API키값" #구글 API 키값
maps = googlemaps.Client(key=my_key)  # my key값 입력

#위도, 경도 변환 함수
# def getLoc(addr):    
#     gmaps = googlemaps.Client(key=my_key)  
#     geocode_result = gmaps.geocode(addr)   
#     n_lat = geocode_result[0]['geometry']['location']['lat']
#     n_lng = geocode_result[0]['geometry']['location']['lng']
#     loc = {'lat':n_lat, 'lng':n_lng}
#     return loc

```

```python
import time


lat = []  #위도
lng = []  #경도

i=0

t1 = time.time()

for address in df['주소']:   
    i = i + 1
    try:
        geo_location = maps.geocode(address)[0].get('geometry')
        lat.append(geo_location['location']['lat'])
        lng.append(geo_location['location']['lng'])
        

    except:
        lat.append('')
        lng.append('')
        print("%d번 인덱스 에러"%(i))


print(time.time() - t1) #지오코딩 총 구동 시간    
```

```python
# 변환한 데이터프레임만들어 출력하기
df2=pd.DataFrame({'건물명':df['건물명'], '주소':df['주소'],'위도':lat,'경도':lng})
#pd.DataFrame([[lat,lng]],columns=['lat','lng'])

df2
```

```python
df2.to_excel('구글_주소_위도_경도.xlsx')
```
